```python
def func_list_flatten(nested_list):
    flattened = []
    for item in nested_list:
        if isinstance(item, list):
            flattened.extend(func_list_flatten(item))
        else:
            flattened.append(item)
    return flattened
```

```python
chaos_list = [list(range(1,21)), list(range(31,45)), [list(range(0,11)), [list(range(12,17))]]]
chaos_list
```

```python
func_list_flatten(chaos_list)
```
