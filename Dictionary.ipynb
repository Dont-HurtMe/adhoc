# Filter Value

```python
data_dict = [{'a': 'AB', 'b': 'SAD', 'list': []}, {'a': 'AB', 'b':'ACCC', 'c':'ad'}, {'a':'cd', 'd':'the', 'c':'HOT'}]
```

```python
def func_filter_value_list_dict(key, value, list_dict_input):
    list_dict_output = []
    for row in list_dict_input : 
        condition_row = row[key] if row[key] is value else None
        if condition_row is not None :
            list_dict_output.append(row)
    return list_dict_output
```

```python
func_filter_value_list_dict('a', 'AB', data_dict)
```
