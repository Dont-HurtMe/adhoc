```python
import threading
import numpy 
from tqdm.notebook import tqdm
```

```python
def split_batch(data, num_batches):
    n_per_batch = len(data) / num_batches
    return [data[int(i*n_per_batch):int((i+1)*n_per_batch)] for i in range(num_batches)]
```

```python
alphabet = [chr(i) for i in range(97, 123)]
alphabet
```

```python
data_batchs = split_batch(alphabet, 5)
print('length per batch: ', [len(i) for i in data_batchs], '\ndata: ', data_batchs)
```

```python
thread_result = []
def func_plus2(data):
    for elm in tqdm(data) : 
        thread_result.append(f"""{elm} <add thread>""")
```

```python
threads = []
for index, data in enumerate(data_batchs):
    thread = threading.Thread(target=func_plus2, args=(data, ))
    thread.start()
    threads.append(thread)
for i, thread in enumerate(threads):
    print(f"start: {i}")
    thread.join()
    print(f"end  : {i}")
```
